# Simple explanation of each column in the dataset:

- Suburb: The neighborhood where the property is located.
- Rooms: The total number of rooms in the property.
- Type: The type of property (e.g., house, unit, townhouse).
- Method: The method of sale (e.g., auction, private sale).
- SellerG: The name of the real estate agent or agency selling the property.
- Distance: The distance of the property from the central business district (CBD) in kilometers.
- Bedroom: The number of bedrooms in the property.
- Bathroom: The number of bathrooms in the property.
- Car: The number of car parking spaces available.
- Landsize: The size of the land in square meters.
- BuildingArea: The size of the building in square meters.
- YearBuilt: The year the property was built.
- Regionname: The region where the property is located.
- Propertycount: The total number of properties in the suburb.
- Price: The sale price of the property.

## Column Name : Type
- h : House - a standalone home.
- t : Townhouse - a multi-floor home sharing walls with others but with its own entrance.
- u : Unit - an apartment or flat in a building.


## Column Name : Method
- SS : Sold after an auction.
- PI : Didn't sell at auction, still available.
- S : Sold before or without an auction.
- SN : Sold, but the price wasn't shared.

# Train Test Split

```python
x = df.drop(columns=['Price'])
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.3,random_state=100)
```

```python
x_train.shape
```

```python
y_train.shape
```

## Standard Scaling :

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
scalar.fit(x_train)

x_train=scalar.transform(x_train)
x_test=scalar.transform(x_test)
```

```python

from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
import numpy as np

# Define the model
model = RandomForestRegressor()

# Define the parameter grid
param_distributions = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

# Define the RandomizedSearchCV  # Defie the Randomized SearchCV
random_search = RandomizedSearchCV(  # Usong random search CV
    estimator=model,
    param_distributions=param_distributions,
    n_iter=50,  # Number of random combinations to try  # Number of random combination to try
    scoring='neg_mean_squared_error',  # Evaluation metric   # Evalution metric
    cv=3,  # 3-fold cross-validation   # 3-folds cross-validation
    verbose=2,  # Verbosity level  # Verbosity level
    random_state=42,
    n_jobs=-1  # Use all available cores  # Use all available cores
)

# Fit the model
random_search.fit(x_train, y_train)

# Best parameters and score
print(f"Best parameters found: {random_search.best_params_}")
print(f"Best score: {random_search.best_score_}")
```
